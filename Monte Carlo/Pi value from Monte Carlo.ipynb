```python
def pivalue(n):
    
    xpoints = np.random.random(n)
    ypoints = np.random.random(n)
    
    inside = 0
    
    for i in range(n):
        
        if (xpoints[i]**2+ ypoints[i]**2)<=1:
            inside+=1
    
    pi = (4*inside)/n
    
    return pi
            
            
```

```python
N = np.arange(100,10000,100)
```

```python
pi_collect=[]

for i in range(np.size(N)):
    pi_collect.append(pivalue(N[i]))
```

```python
pi = np.ones_like(N)*np.pi
```

```python
plt.figure(figsize=(8,4))
plt.plot(N,pi_collect,"b")
plt.plot(N,pi,"r--",label="$\pi$ value")
plt.xlabel("Number of points")
plt.ylabel("$\pi$ calculated")
plt.legend()
plt.title("$\pi$ value calculated using Monte Carlo for various steps")
plt.grid()
plt.savefig("pivsn.pdf")
plt.show()
```

```python
fig, axs = plt.subplots(2,figsize=(8,6))
fig.suptitle("$\pi$ value calculated using Monte Carlo for various steps")
axs[0].plot(N,pi_collect,"g")
axs[0].plot(N,pi,"--",label="$\pi$ value")
axs[0].grid(True)
axs[1].loglog(N,np.abs(pi-pi_collect),"k-")
axs[1].grid(True)
```

```python
def pi_hist(N,trials):
    
    pi = []
    
    for n in range(trials):
        pi.append(pivalue(N))
    
    pimean = np.mean(pi)
    plt.figure(figsize=[6,5])
    Bins=plt.hist(pi,bins='auto',label="$\pi = ${:3f}".format(pimean))
    
    
    plt.xlabel("$\pi$")
    plt.ylabel("Counts")
    
    plt.title("Histogram of pi calculated from N = {} points and {} trials".format(N,trials))
    plt.legend()
    plt.grid()
    plt.savefig("pihist{}{}.pdf".format(N,trials))
    plt.show()
```

```python
pi_hist(100,5000)
```

```python

pi_hist(100,100)
pi_hist(100,500)
pi_hist(100,1000)
```
